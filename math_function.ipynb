# Math functions

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math as log

# Logarithm
x= np.arange(1,10,1)
y= []
print(x)

for i in x:
    log_y = log.log(i, 2)
    y.append(log_y)
print(y)

plt.figure(figsize=(8,8))
plt.plot(x,y, color='red')
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.grid()
plt.show()
```
